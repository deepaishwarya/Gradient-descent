# Gradient Descent Algorithm implementation

```python
def gradient_descent(x,y):
    m_curr=c_curr=0;
    r=1000;#no. of iterations
    n=len(x);
    l=0.001;
    cost_arr=[]
    for i in range(r):
        y_predicted = m_curr*x+c_curr;
       
        cost=1/(2*n)*sum([val**2 for val in (y-y_predicted)]);
        
        cost_arr.append(cost)
        cd=1/n*(sum(y-y_predicted));#derivative of the parameter c
        md=1/n*(sum(x*(y-y_predicted)));#derivative of the parameter m
        c_curr=c_curr-l*cd;#updating the value of parameter c_curr
        m_curr=m_curr-l*md;#updating the value of parameter m_curr
        cost_arr=[]
        cost_arr.append(cost)
        print("m_curr={},c_curr={},cost={},i={}".format(m_curr,c_curr,cost,i))
    
```

```python
gradient_descent(X_data,y_data)
```
